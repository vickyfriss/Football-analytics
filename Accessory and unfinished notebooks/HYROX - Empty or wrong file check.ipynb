```python
import os
import pandas as pd

BASE_FOLDER = "Datasets/Hyrox/"

deleted_files = []
checked_files = 0

print("\n==============================")
print("STRICT HYROX DATA VALIDATION")
print("==============================\n")

for season in os.listdir(BASE_FOLDER):
    season_path = os.path.join(BASE_FOLDER, season)

    if not os.path.isdir(season_path):
        continue

    for file in os.listdir(season_path):
        if not file.endswith(".csv"):
            continue

        file_path = os.path.join(season_path, file)
        checked_files += 1

        try:
            df = pd.read_csv(file_path)
        except Exception:
            print(f"❌ Could not read {file} — deleting")
            os.remove(file_path)
            deleted_files.append(file_path)
            continue

        # Empty file
        if df.empty:
            print(f"❌ Empty file deleted: {file}")
            os.remove(file_path)
            deleted_files.append(file_path)
            continue

        # Missing Gender column
        if "Gender" not in df.columns:
            print(f"❌ No Gender column: {file}")
            os.remove(file_path)
            deleted_files.append(file_path)
            continue

        gender_counts = df["Gender"].value_counts()

        # ----------------------------
        # RULE 1 — Missing gender
        # ----------------------------
        if len(gender_counts) < 2:
            print(f"❌ Missing one gender in {file} ({list(gender_counts.index)})")
            os.remove(file_path)
            deleted_files.append(file_path)
            continue

        suspicious = False

        # ----------------------------
        # RULE 2 — Exact 100 multiples
        # ----------------------------
        for gender, count in gender_counts.items():
            if count % 100 == 0:
                print(f"❌ {file} — {gender} has exactly {count} rows")
                suspicious = True
                break

        if suspicious:
            os.remove(file_path)
            deleted_files.append(file_path)
            continue

        print(f"✓ OK: {file}")

print("\n==============================")
print(f"Checked files: {checked_files}")
print(f"Deleted files: {len(deleted_files)}")
print("==============================")
```

```python
import os
import pandas as pd

BASE_FOLDER = "Datasets/Hyrox/"

checked_files = 0
deleted_files = []

print("\n==============================")
print("DELETE HYROX/HYROX_PRO FILES WITH ONLY 'All' GENDER")
print("==============================\n")

for season in os.listdir(BASE_FOLDER):
    season_path = os.path.join(BASE_FOLDER, season)

    if not os.path.isdir(season_path):
        continue

    for file in os.listdir(season_path):
        if not file.endswith(".csv"):
            continue

        # Only target HYROX or HYROX_PRO divisions
        if "HYROX" not in file.upper() and "HYROX_PRO" not in file.upper():
            continue

        file_path = os.path.join(season_path, file)
        checked_files += 1

        try:
            df = pd.read_csv(file_path)
        except Exception:
            print(f"❌ Could not read {file} — skipping")
            continue

        # Skip empty files
        if df.empty:
            continue

        if "Gender" not in df.columns:
            print(f"❌ No Gender column in {file} — skipping")
            continue

        gender_counts = df["Gender"].value_counts()

        # Delete if only 'All' exists
        if len(gender_counts) == 1 and "All" in gender_counts.index:
            print(f"❌ Deleting {file} — only 'All' gender")
            os.remove(file_path)
            deleted_files.append(file_path)

print("\n==============================")
print(f"Checked HYROX/HYROX_PRO files: {checked_files}")
print(f"Deleted files: {len(deleted_files)}")
print("==============================")
```
