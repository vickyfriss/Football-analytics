```python
import http.client
import json
import pandas as pd

conn = http.client.HTTPSConnection("v3.football.api-sports.io")

headers = {
    'x-apisports-key': API_KEY
}

conn.request("GET", "/leagues", headers=headers)
res = conn.getresponse()
data = res.read()

# Parse JSON
json_data = json.loads(data.decode("utf-8"))

# Convert to DataFrame
leagues_df = pd.json_normalize(json_data["response"])
```

```python
import pandas as pd

# explode seasons list â†’ one row per season
leagues_seasons = leagues_df.explode("seasons").reset_index(drop=True)

# normalize season dict into columns
season_cols = pd.json_normalize(leagues_seasons["seasons"])

# combine with league info
leagues_seasons = pd.concat(
    [leagues_seasons.drop(columns=["seasons"]), season_cols],
    axis=1
)

leagues_seasons
```

```python
uruguay_leagues = leagues_seasons[
    (leagues_seasons["country.name"] == "Uruguay") &
    (leagues_seasons["year"] >= 2023) & (leagues_seasons["year"] < 2026) &
    (leagues_seasons["league.id"].isin([268, 270]))
]

uruguay_leagues
```

```python
import requests
import pandas as pd
import time


headers = {"x-apisports-key": API_KEY}
BASE = "https://v3.football.api-sports.io"

def get_fixtures(league_id, season):
    url = f"{BASE}/fixtures"
    params = {"league": league_id, "season": season}
    r = requests.get(url, headers=headers, params=params)
    return r.json()["response"]

# collect all fixture IDs

fixtures_all = []

for _, row in uruguay_leagues.iterrows():
    league_id = row["league.id"]
    season = row["year"]

    print(f"Downloading fixtures â†’ league {league_id}, season {season}")
    fixtures = get_fixtures(league_id, season)
    fixtures_all.extend(fixtures)

    time.sleep(1)   # rate limit protection

fixtures_df = pd.json_normalize(fixtures_all)
fixture_ids = fixtures_df["fixture.id"].tolist()
```

```python
# download events for every match

def get_events(fixture_id):
    url = f"{BASE}/fixtures/events"
    params = {"fixture": fixture_id}
    r = requests.get(url, headers=headers, params=params)
    return r.json()["response"]

events_all = []

for i, fid in enumerate(fixture_ids):
    print(f"Events {i+1}/{len(fixture_ids)}")

    events = get_events(fid)
    for e in events:
        e["fixture_id"] = fid

    events_all.extend(events)

    time.sleep(1)  # IMPORTANT (100 requests/day)

events_df = pd.json_normalize(events_all)
```

```python
# also get lineups

def get_lineups(fixture_id):
    url = f"{BASE}/fixtures/lineups"
    params = {"fixture": fixture_id}
    r = requests.get(url, headers=headers, params=params)
    return r.json()["response"]

lineups_all = []

for fid in fixture_ids:
    data = get_lineups(fid)
    lineups_all.append({
        "fixture_id": fid,
        "data": data
    })
    time.sleep(1)

lineups_df = pd.json_normalize(lineups_all)
```

```python
import os
import json
import http.client
import time

# Helper functions

HEADERS = {
    "x-apisports-key": API_KEY
}


def make_fixture_folder(fixture_id):
    path = os.path.join(BASE_PATH, str(fixture_id))
    os.makedirs(path, exist_ok=True)
    return path


def file_exists(fixture_id, filename):
    path = os.path.join(BASE_PATH, str(fixture_id), filename)
    return os.path.exists(path)


def save_json(data, fixture_id, filename):
    path = os.path.join(BASE_PATH, str(fixture_id), filename)
    with open(path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2)

def valid_json(path):

    if not os.path.exists(path):
        return False

    if os.path.getsize(path) == 0:
        return False

    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)

        if data.get("errors"):
            return False

        return True

    except:
        return False

def file_complete(fixture_id, filename):
    path = os.path.join(make_fixture_folder(fixture_id), filename)
    return valid_json(path)

def api_get_safe(endpoint):

    data = api_get(endpoint)

    if data.get("errors"):
        print("\nAPI LIMIT OR ERROR")
        print(data["errors"])
        raise SystemExit("Stopping downloader")

    return data
```

```python
# API request function

def api_get(endpoint):
    conn = http.client.HTTPSConnection("v3.football.api-sports.io")
    conn.request("GET", endpoint, headers=HEADERS)
    res = conn.getresponse()
    data = res.read()
    return json.loads(data.decode("utf-8"))
```

```python
def download_fixture_full(fixture_id):

    print(f"\nDownloading fixture {fixture_id}")

    folder = make_fixture_folder(fixture_id)

    # ---------------- EVENTS ----------------
    if not file_complete(fixture_id, "events.json"):
        data = api_get_safe(f"/fixtures/events?fixture={fixture_id}")
        save_json(data, fixture_id, "events.json")
        print("   events saved")

    # ---------------- PLAYERS ----------------
    if not file_complete(fixture_id, "players.json"):
        data = api_get_safe(f"/fixtures/players?fixture={fixture_id}")
        save_json(data, fixture_id, "players.json")
        print("   players saved")

    # ---------------- LINEUPS ----------------
    if not file_complete(fixture_id, "lineups.json"):
        data = api_get_safe(f"/fixtures/lineups?fixture={fixture_id}")
        save_json(data, fixture_id, "lineups.json")
        print("   lineups saved")

    # ---------------- STATISTICS ----------------
    if not file_complete(fixture_id, "statistics.json"):
        data = api_get_safe(f"/fixtures/statistics?fixture={fixture_id}")
        save_json(data, fixture_id, "statistics.json")
        print("   statistics saved")

    print("   fixture complete")
```

```python
# FULL PARSER SCRIPT

import os
import json
import pandas as pd

BASE_PATH = r"Datasets\Uruguay\fixtures"
OUTPUT_PATH = r"Datasets\Uruguay"

# --------------------------------------------------
# Helpers
# --------------------------------------------------

def load_json(path):
    if not os.path.exists(path):
        return None
    with open(path, encoding="utf-8") as f:
        return json.load(f)

def list_fixture_dirs():
    return [
        d for d in os.listdir(BASE_PATH)
        if os.path.isdir(os.path.join(BASE_PATH, d))
    ]

# --------------------------------------------------
# EVENTS
# --------------------------------------------------

def parse_events(fixture_id, data):

    rows = []

    for ev in data.get("response", []):
        row = {
            "fixture_id": fixture_id,
            "time_elapsed": ev.get("time", {}).get("elapsed"),
            "time_extra": ev.get("time", {}).get("extra"),
            "team_id": ev.get("team", {}).get("id"),
            "team_name": ev.get("team", {}).get("name"),
            "player_id": ev.get("player", {}).get("id"),
            "player_name": ev.get("player", {}).get("name"),
            "assist_id": ev.get("assist", {}).get("id"),
            "assist_name": ev.get("assist", {}).get("name"),
            "event_type": ev.get("type"),
            "event_detail": ev.get("detail"),
            "comments": ev.get("comments")
        }
        rows.append(row)

    return rows

# --------------------------------------------------
# PLAYER MATCH STATS
# --------------------------------------------------

def parse_players(fixture_id, data):

    rows = []

    for team in data.get("response", []):
        team_id = team["team"]["id"]
        team_name = team["team"]["name"]

        for p in team.get("players", []):
            stats = p["statistics"][0]

            row = {
                "fixture_id": fixture_id,
                "team_id": team_id,
                "team_name": team_name,
                "player_id": p["player"]["id"],
                "player_name": p["player"]["name"],
                "position": p["statistics"][0]["games"]["position"],
                "minutes": stats["games"]["minutes"],
                "rating": stats["games"]["rating"],
                "goals": stats["goals"]["total"],
                "assists": stats["goals"]["assists"],
                "shots_total": stats["shots"]["total"],
                "shots_on": stats["shots"]["on"],
                "passes_total": stats["passes"]["total"],
                "passes_accuracy": stats["passes"]["accuracy"],
                "tackles": stats["tackles"]["total"],
                "duels_total": stats["duels"]["total"],
                "duels_won": stats["duels"]["won"],
                "dribbles_attempts": stats["dribbles"]["attempts"],
                "dribbles_success": stats["dribbles"]["success"]
            }

            rows.append(row)

    return rows

# --------------------------------------------------
# LINEUPS
# --------------------------------------------------

def parse_lineups(fixture_id, data):

    rows = []

    for team in data.get("response", []):
        team_id = team["team"]["id"]
        team_name = team["team"]["name"]
        formation = team.get("formation")

        for player in team.get("startXI", []):
            rows.append({
                "fixture_id": fixture_id,
                "team_id": team_id,
                "team_name": team_name,
                "formation": formation,
                "player_id": player["player"]["id"],
                "player_name": player["player"]["name"],
                "lineup_type": "starter"
            })

        for player in team.get("substitutes", []):
            rows.append({
                "fixture_id": fixture_id,
                "team_id": team_id,
                "team_name": team_name,
                "formation": formation,
                "player_id": player["player"]["id"],
                "player_name": player["player"]["name"],
                "lineup_type": "sub"
            })

    return rows

# --------------------------------------------------
# TEAM MATCH STATS
# --------------------------------------------------

def parse_team_stats(fixture_id, data):

    rows = []

    for team in data.get("response", []):
        team_id = team["team"]["id"]
        team_name = team["team"]["name"]

        row = {
            "fixture_id": fixture_id,
            "team_id": team_id,
            "team_name": team_name
        }

        for stat in team.get("statistics", []):
            key = stat["type"].lower().replace(" ", "_").replace("%", "pct")
            row[key] = stat["value"]

        rows.append(row)

    return rows

# --------------------------------------------------
# MAIN PARSER LOOP
# --------------------------------------------------

events_all = []
players_all = []
lineups_all = []
teams_all = []

fixture_dirs = list_fixture_dirs()

print("Fixtures found:", len(fixture_dirs))

for fid in fixture_dirs:

    folder = os.path.join(BASE_PATH, fid)

    events = load_json(os.path.join(folder, "events.json"))
    players = load_json(os.path.join(folder, "players.json"))
    lineups = load_json(os.path.join(folder, "lineups.json"))
    stats = load_json(os.path.join(folder, "statistics.json"))

    if events:
        events_all.extend(parse_events(fid, events))

    if players:
        players_all.extend(parse_players(fid, players))

    if lineups:
        lineups_all.extend(parse_lineups(fid, lineups))

    if stats:
        teams_all.extend(parse_team_stats(fid, stats))

print("Parsing complete")

# --------------------------------------------------
# SAVE TABLES
# --------------------------------------------------

pd.DataFrame(events_all).to_csv(
    os.path.join(OUTPUT_PATH, "events.csv"), index=False
)

pd.DataFrame(players_all).to_csv(
    os.path.join(OUTPUT_PATH, "player_match_stats.csv"), index=False
)

pd.DataFrame(lineups_all).to_csv(
    os.path.join(OUTPUT_PATH, "lineups.csv"), index=False
)

pd.DataFrame(teams_all).to_csv(
    os.path.join(OUTPUT_PATH, "team_match_stats.csv"), index=False
)

print("CSV tables saved")
```
