## Hypothesis testing

```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

# Assume your DataFrame is called conversion_table
# It has columns: ['Shot_height_zone', 'Shot_width_zone', 'converted', 'total', 'conversion_rate']

# Aggregate over width zones to get total converted and total attempts per height zone
agg = conversion_table.groupby('Shot_height_zone').agg({
    'converted': 'sum',
    'total': 'sum'
}).reset_index()

# Convert to dictionary for easy access
zones = {row['Shot_height_zone']: (row['converted'], row['total']) for idx, row in agg.iterrows()}

# Function to test if B > other zone
def test_middle_vs(zone_name):
    count = [zones['B'][0], zones[zone_name][0]]
    nobs = [zones['B'][1], zones[zone_name][1]]
    stat, pval = proportions_ztest(count, nobs, alternative='larger')  # B > other
    return stat, pval

# Test B vs A
stat_BA, pval_BA = test_middle_vs('A')
print(f"B vs A: z = {stat_BA:.3f}, p-value = {pval_BA:.4f}")

# Test B vs C
stat_BC, pval_BC = test_middle_vs('C')
print(f"B vs C: z = {stat_BC:.3f}, p-value = {pval_BC:.4f}")

# Optional: check which zone actually has the highest conversion rate
agg['conversion_rate'] = agg['converted'] / agg['total']
best_zone = agg.loc[agg['conversion_rate'].idxmax()]
print(f"\nHighest conversion rate is in zone {best_zone['Shot_height_zone']} with {best_zone['conversion_rate']:.3f}")
```

```python
import pandas as pd

# Copy original data
df = PL_UCL_22_23.copy()

# Function to assign off-target shots to nearest zone
def assign_off_target(row):
    h = row['Shot_height_zone']
    w = row['Shot_width_zone']
    
    # Special corners
    if h == 'O' and w == 0:
        return pd.Series({'Adjusted_height': 'A', 'Adjusted_width': 1})
    if h == 'O' and w == 4:
        return pd.Series({'Adjusted_height': 'A', 'Adjusted_width': 3})
    
    # Vertical off-target: assign top 'A', keep width if 1,2,3
    if h == 'O' and w in [1,2,3]:
        return pd.Series({'Adjusted_height': 'A', 'Adjusted_width': w})
    
    # Horizontal off-targets
    if w == 0 and h in ['A','B','C']:
        return pd.Series({'Adjusted_height': h, 'Adjusted_width': 1})
    if w == 4 and h in ['A','B','C']:
        return pd.Series({'Adjusted_height': h, 'Adjusted_width': 3})
    
    # Otherwise, keep original
    return pd.Series({'Adjusted_height': h, 'Adjusted_width': w})

# Apply the off-target adjustment
df[['Adjusted_height', 'Adjusted_width']] = df.apply(assign_off_target, axis=1)

# Aggregate by adjusted zones
conversion_table = (
    df.groupby(['Adjusted_height', 'Adjusted_width'])
    .agg(
        converted=('Goal', 'sum'),
        saved=('Saved', 'sum'),
        missed=('Off_target', 'sum'),
        total=('Goal', 'count')
    )
    .reset_index()
)

# Compute conversion rate
conversion_table['conversion_rate'] = conversion_table['converted'] / conversion_table['total']

conversion_table
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.patches import Rectangle
import matplotlib.colors as mcolors

# Load goal image
img = mpimg.imread("Images and others/soccer goal-2.jpg")

fig, ax = plt.subplots(figsize=(8, 4))
ax.imshow(img, extent=[0, img.shape[1], 0, img.shape[0]], zorder=0)

# Goal pixel bounds
goal_x_start_px = 85
goal_x_end_px = 322
goal_y_start_px = 40
goal_y_end_px = 145

# Map zones to goal coordinates
x_scale = (goal_x_end_px - goal_x_start_px) / 7.32
y_scale = (goal_y_end_px - goal_y_start_px) / 2.44

width_bins = {1: (0, 2.44), 2: (2.44, 4.88), 3: (4.88, 7.32)}
height_bins = {'A': (1.63, 2.44), 'B': (0.81, 1.63), 'C': (0, 0.81)}  # top to bottom

# Gradient traffic-light colormap: red -> yellow -> green
cmap = mcolors.LinearSegmentedColormap.from_list('traffic_light', ['red', 'yellow', 'green'])
norm = mcolors.Normalize(vmin=0, vmax=1)

# Overlay zones with colored rectangles and conversion rate text
for _, row in conversion_table.iterrows():
    wz = row["Adjusted_width"]
    hz = row["Adjusted_height"]

    if wz not in width_bins or hz not in height_bins:
        continue

    x0, x1 = width_bins[wz]
    y0, y1 = height_bins[hz]

    # Convert to pixel coordinates
    px0 = goal_x_start_px + x0 * x_scale
    px1 = goal_x_start_px + x1 * x_scale
    py0 = goal_y_start_px + y0 * y_scale
    py1 = goal_y_start_px + y1 * y_scale

    # Rectangle colored by conversion rate
    ax.add_patch(
        Rectangle(
            (px0, py0),
            px1 - px0,
            py1 - py0,
            facecolor=cmap(norm(row['conversion_rate'])),
            alpha=0.65,
            edgecolor="black",
            zorder=2
        )
    )

    # Add text: conversion rate in %
    ax.text(
        (px0 + px1)/2,  # center x
        (py0 + py1)/2,  # center y
        f"{row['conversion_rate']*100:.1f}%",  # show as percentage
        color='black' if row['conversion_rate'] < 0.5 else 'white',  # contrast text
        ha='center',
        va='center',
        fontsize=10,
        fontweight='bold',
        zorder=3
    )

ax.axis("off")
plt.title("Penalty Conversion Rate Heatmap (Traffic Light Gradient)")
plt.show()
```

```python
import pandas as pd

# Map width zones to Left / Middle / Right
def map_horizontal(w):
    if w == 1:
        return 'Left'
    elif w == 2:
        return 'Middle'
    elif w == 3:
        return 'Right'
    else:
        return 'Other'  # in case there are out-of-bound widths

conversion_table['H_zone'] = conversion_table['Adjusted_width'].apply(map_horizontal)

# Group by horizontal zone only
agg_horizontal = (
    conversion_table.groupby('H_zone')
    .agg(
        converted=('converted', 'sum'),
        saved=('saved', 'sum'),
        missed=('missed', 'sum'),
        total=('total', 'sum')
    )
    .reset_index()
)

# Compute conversion rate
agg_horizontal['conversion_rate'] = agg_horizontal['converted'] / agg_horizontal['total']

agg_horizontal
```
