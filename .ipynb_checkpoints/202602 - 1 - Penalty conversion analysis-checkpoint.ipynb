```python
conversion_table = (
    PL_UCL_22_23
    .groupby(["Shot_height_zone", "Shot_width_zone"])
    .agg(
        converted=("Goal", "sum"),
        total=("Goal", "count")
    )
)

conversion_table["conversion_rate"] = (
    conversion_table["converted"] / conversion_table["total"]
)

conversion_table
```
