```python
all_matches = []

for _, row in women_comps.iterrows():
    matches = sb.matches(
        competition_id=row["competition_id"],
        season_id=row["season_id"]
    )
    all_matches.append(matches)

matches_df = pd.concat(all_matches, ignore_index=True)
```

## All events, passes only

```python
from pathlib import Path
import pandas as pd

CACHE_DIR = Path("statsbomb_cache/events")
CACHE_DIR.mkdir(parents=True, exist_ok=True)


def load_events(match_id):
    cache_file = CACHE_DIR / f"{match_id}.parquet"

    if cache_file.exists():
        return pd.read_parquet(cache_file)

    events = sb.events(match_id=match_id)
    events.to_parquet(cache_file)
    return events
```

```python
all_passes = []

for match_id in matches_df["match_id"]:
    events = load_events(match_id)
    passes = events[events["type"] == "Pass"]
    all_passes.append(passes)

passes_df = pd.concat(all_passes, ignore_index=True)
```

## Feature engineering

```python
def prepare_pass_features(df_passes):
    df = df_passes.copy()  # work on a copy to avoid modifying original

    # Pitch dimensions
    PITCH_LENGTH = 120
    PITCH_WIDTH = 80
    GOAL_X = PITCH_LENGTH
    GOAL_Y = PITCH_WIDTH / 2

    # Extract start / end locations
    df["start_x"] = df["location"].str[0]
    df["start_y"] = df["location"].str[1]
    df["end_x"] = df["pass_end_location"].str[0]
    df["end_y"] = df["pass_end_location"].str[1]

    # Remove Injury Clearance passes
    df = df[df["pass_outcome"].ne("Injury Clearance")].copy()

    # Define pitch bins (10-yard grid)
    x_bins = np.linspace(0, PITCH_LENGTH, 13)
    y_bins = np.linspace(0, PITCH_WIDTH, 9)

    df["start_x_zone"] = pd.cut(df["start_x"], bins=x_bins, labels=False, include_lowest=True)
    df["start_y_zone"] = pd.cut(df["start_y"], bins=y_bins, labels=False, include_lowest=True)
    df["end_x_zone"] = pd.cut(df["end_x"], bins=x_bins, labels=False, include_lowest=True)
    df["end_y_zone"] = pd.cut(df["end_y"], bins=y_bins, labels=False, include_lowest=True)

    # Clip zones to max bin
    df["start_x_zone"] = df["start_x_zone"].clip(upper=len(x_bins)-2)
    df["start_y_zone"] = df["start_y_zone"].clip(upper=len(y_bins)-2)
    df["end_x_zone"] = df["end_x_zone"].clip(upper=len(x_bins)-2)
    df["end_y_zone"] = df["end_y_zone"].clip(upper=len(y_bins)-2)

    # Drop rows with missing start zones
    df = df.dropna(subset=["start_x_zone", "start_y_zone"]).copy()
    df["start_x_zone"] = df["start_x_zone"].astype(int)
    df["start_y_zone"] = df["start_y_zone"].astype(int)

    # Combine x/y zones into single square
    df["start_square"] = df["start_x_zone"].astype(str) + "_" + df["start_y_zone"].astype(str)

    # Create full list of all possible squares
    all_squares = [f"{i}_{j}" for i in range(12) for j in range(8)]  # 12 x 8 grid
    start_square_dummies = pd.get_dummies(df["start_square"], prefix="sq").reindex(
        columns=[f"sq_{sq}" for sq in all_squares], fill_value=0
    )
    df = pd.concat([df, start_square_dummies], axis=1)

    # Straight-line pass distance
    df["pass_distance"] = np.sqrt(
        (df["end_x"] - df["start_x"])**2 +
        (df["end_y"] - df["start_y"])**2
    )

    # Distance from ball to opponent goal
    df["start_dist_goal"] = np.sqrt((GOAL_X - df["start_x"])**2 + (GOAL_Y - df["start_y"])**2)
    df["end_dist_goal"] = np.sqrt((GOAL_X - df["end_x"])**2 + (GOAL_Y - df["end_y"])**2)

    # How much closer the pass moves the ball towards goal
    df["progression"] = df["start_dist_goal"] - df["end_dist_goal"]

    # Progressive pass (>10 yards closer to goal)
    df["progressive"] = (df["progression"] > 10).astype(int)

    # Combined difficulty features
    df["progressive_opp_half"] = df["progressive"] * (df["start_x_zone"] >= 6).astype(int)
    df["central_progressive"] = df["progressive"] * (
        (df["start_y_zone"] >= 3) & (df["start_y_zone"] <= 5)
    ).astype(int)

    # Crosses under pressure
    df["under_pressure"] = df["under_pressure"].fillna(False).astype(int)
    df["cross"] = df["pass_cross"].fillna(False).astype(int)
    df["cross_under_pressure"] = df["cross"] * df["under_pressure"]

    # Long passes under pressure
    df["long_pass"] = (df["pass_distance"] > 25).astype(int)
    df["very_long_pass"] = (df["pass_distance"] > 40).astype(int)
    df["long_under_pressure"] = df["long_pass"] * df["under_pressure"]

    # High balls / through balls / switches
    df["high_pass"] = (df["pass_height"] == "High Pass").astype(int)
    df["through_ball"] = df["pass_through_ball"].fillna(False).astype(int)
    df["switch"] = df["pass_switch"].fillna(False).astype(int)

    # Play pattern one-hot encoding (Regular Play as baseline)
    play_pattern_dummies = pd.get_dummies(df["play_pattern"], prefix="pp", drop_first=True)
    df = pd.concat([df, play_pattern_dummies], axis=1)

    # Completed pass flag
    df["completed"] = df["pass_outcome"].isna().astype(int)

    # Final feature list for xP model
    FEATURES = [
        "pass_distance", "progression", "start_dist_goal", "under_pressure",
        "high_pass", "through_ball", "cross", "switch",
        "long_under_pressure", "very_long_pass", "progressive_opp_half",
        "cross_under_pressure", "central_progressive"
    ]
    FEATURES += list(play_pattern_dummies.columns)
    FEATURES += list(start_square_dummies.columns)

    df = df.copy()  # defragment
    X = df[FEATURES]
    y = df["completed"]

    return df, X, y, FEATURES
```

```python
df, X, y, FEATURES = prepare_pass_features(passes_df)
```

## Train global xP model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
import pandas as pd
import numpy as np

# --------------------------------------------------
# Fit expected pass completion (xP) model
# --------------------------------------------------

# Simple logistic regression is enough here.
# The goal is interpretability, not squeezing every last % of performance.
model = LogisticRegression(max_iter=1000)

model.fit(X, y)

# --------------------------------------------------
# Check coefficients to see if they make football sense
# --------------------------------------------------

coef_table = pd.DataFrame({
    "feature": FEATURES,
    "coef": model.coef_[0]
}).sort_values("coef")

coef_table
```

```python
# --------------------------------------------------
# Generate expected completion probability (xP)
# --------------------------------------------------

df["xP"] = model.predict_proba(X)[:, 1]

# Basic discrimination check
auc = roc_auc_score(y, df["xP"])
print(f"AUC: {auc:.3f}")
```

```python
def aggregate_xP(df_feat, X, y, model):
    df = df_feat.copy()

    # -------------------------------
    # Ensure X has all features model expects
    # -------------------------------
    model_features = model.feature_names_in_  # sklearn keeps this
    for col in model_features:
        if col not in X.columns:
            X[col] = 0

    # Reorder columns to match training
    X = X[model_features]

    # -------------------------------
    # Predict xP for each pass
    # -------------------------------
    df["xP"] = model.predict_proba(X)[:, 1]

    # -------------------------------
    # Aggregate per team
    # -------------------------------
    team_stats = df.groupby("team_name").agg(
        total_passes=("completed", "count"),
        completed_passes=("completed", "sum"),
        total_xP=("xP", "sum")
    ).reset_index()

    # Calculate difference between actual and expected
    team_stats["diff"] = team_stats["completed_passes"] - team_stats["total_xP"]

    # -------------------------------
    # Aggregate per player
    # -------------------------------
    player_stats = df.groupby("player_name").agg(
        total_passes=("completed", "count"),
        completed_passes=("completed", "sum"),
        total_xP=("xP", "sum")
    ).reset_index()

    # Calculate difference between actual and expected
    player_stats["diff"] = player_stats["completed_passes"] - player_stats["total_xP"]

    return team_stats, player_stats
```

```python
# Euros final aggregates
team_euros, player_euros = aggregate_xP(df_euros_feat, X_euros, y_euros, model)

# Worlds final aggregates
team_world, player_world = aggregate_xP(df_world_feat, X_world, y_world, model)
```
