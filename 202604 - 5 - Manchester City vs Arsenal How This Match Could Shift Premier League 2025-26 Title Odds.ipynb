# Title Odds Visualisation - Manchester City vs Arsenal  

**Purpose:** Create a clean visualisation of Premier League title probabilities under different match outcomes (pre-match, Arsenal win, draw, Manchester City win).  
**Data:** Simulation model outputs (title probabilities for each scenario).  
**Methods:** Scenario comparison, grouped bar chart visualisation.  
**Tools:** Python, Matplotlib, NumPy  
**Author:** [Victoria Friss de Kereki](https://www.linkedin.com/in/victoria-friss-de-kereki/)  
**Article:** [Manchester City vs Arsenal: How This Match Could Shift Premier League 2025/26 Title Odds](https://medium.com/@vickyfrissdekereki/manchester-city-vs-arsenal-how-this-match-could-shift-premier-league-2025-26-title-odds-7a838e44c197)   

**Created and Last Updated:** `17/04/2026`  

> This script generates a grouped bar chart comparing title probabilities for Arsenal and Manchester City across different match scenarios.  
>  
> The focus is on presenting probability shifts clearly and intuitively, using minimal styling and direct annotation to highlight differences between outcomes.  

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

# ----------------------------
# Data
# ----------------------------
scenarios = ["Pre-match", "Arsenal Win", "Draw", "Manchester City Win"]

arsenal = [81, 99, 90, 65]
city = [19, 1, 10, 35]

x = np.arange(len(scenarios)) * 1.12

width = 0.28
gap = 0.05

# ----------------------------
# Style
# ----------------------------
plt.style.use("seaborn-v0_8-white")

fig, ax = plt.subplots(figsize=(7.2, 3.25), dpi=180)

fig.patch.set_facecolor("#ffffff")
ax.set_facecolor("#ffffff")

arsenal_color = "#EF0107"
city_color = "#6CABDD"

# ----------------------------
# Bars
# ----------------------------
bars_arsenal = ax.bar(x - (width/2 + gap/2), arsenal, width, color=arsenal_color)
bars_city = ax.bar(x + (width/2 + gap/2), city, width, color=city_color)

# ----------------------------
# HEADER
# ----------------------------
fig.text(
    0.5, 0.965,
    "Manchester City vs Arsenal - Title Chances by Match Outcome",
    ha="center",
    fontsize=13,
    fontweight="bold"
)

fig.lines.append(plt.Line2D(
    [0.18, 0.82], [0.935, 0.935],
    transform=fig.transFigure,
    color="#dddddd",
    linewidth=1
))

fig.text(
    0.5, 0.885,
    "How the Premier League 2025/26 title race shifts under different match scenarios",
    ha="center",
    fontsize=9.5,
    color="#666666"
)

# ----------------------------
# X axis
# ----------------------------
ax.set_xticks(x)
ax.set_xticklabels(scenarios, fontsize=10)
ax.tick_params(axis='x', length=0)

# ----------------------------
# Y axis clean
# ----------------------------
ax.set_ylim(0, 105)
ax.set_yticks([])
ax.set_ylabel("")

# ----------------------------
# Frame styling
# ----------------------------
for spine in ["top", "right", "left"]:
    ax.spines[spine].set_visible(False)

ax.spines["bottom"].set_alpha(0.22)

ax.grid(axis="y", linestyle="--", alpha=0.04)

# ----------------------------
# Value labels
# ----------------------------
for bar in list(bars_arsenal) + list(bars_city):
    h = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2,
        h + 1.0,
        f"{int(h)}%",
        ha="center",
        va="bottom",
        fontsize=10
    )

# ----------------------------
# Legend
# ----------------------------
legend_elements = [
    Patch(facecolor=arsenal_color, label="Arsenal"),
    Patch(facecolor=city_color, label="Manchester City")
]

ax.legend(
    handles=legend_elements,
    frameon=False,
    loc="upper right",
    fontsize=9
)

# ----------------------------
# FOOTER (robust, no overlap system)
# ----------------------------

footer_x = 0.995

name_y = 0.04
role_y = 0.008   # fixed spacing (prevents overlap permanently)

fig.text(
    footer_x, name_y,
    "Victoria Friss de Kereki",
    ha="right",
    fontsize=8.6,
    color="#888888"
)

fig.text(
    footer_x, role_y,
    "Sports Analytics & Data Science",
    ha="right",
    fontsize=7.8,
    color="#aaaaaa"
)

# ----------------------------
# Layout control
# ----------------------------
plt.tight_layout(rect=[0, 0.06, 1, 0.86])


import os

# ----------------------------
# Save figure
# ----------------------------
output_dir = "Images and others"
os.makedirs(output_dir, exist_ok=True)

file_path = os.path.join(output_dir, "man_city_vs_arsenal_title_odds.png")

plt.savefig(
    file_path,
    dpi=300,
    bbox_inches="tight",
    facecolor="white"
)


plt.show()
```

```python
import matplotlib.pyplot as plt
import os

# ----------------------------
# Style
# ----------------------------
arsenal_color = "#EF0107"
city_color = "#6CABDD"

output_dir = "Images and others"
os.makedirs(output_dir, exist_ok=True)

# ----------------------------
# Compact, refined card
# ----------------------------
def create_big_card(title, arsenal_val, city_val, filename):
    
    fig = plt.figure(figsize=(2.2, 1.0), dpi=300)
    fig.patch.set_facecolor("white")

    # ----------------------------
    # Title (very subtle + centred)
    # ----------------------------
    plt.figtext(
        0.5, 0.80,
        title,
        ha="center",
        va="center",
        fontsize=5.2,      # ↓ smaller again
        color="#888888"    # ↓ lighter grey
    )

    # Divider
    fig.lines.append(plt.Line2D(
        [0.30, 0.70], [0.74, 0.74],
        transform=fig.transFigure,
        color="#dddddd",
        linewidth=0.5
    ))

    # ----------------------------
    # Team names
    # ----------------------------
    plt.figtext(
        0.37, 0.56,
        "Arsenal",
        ha="center",
        va="center",
        fontsize=6.2,
        color=arsenal_color,
        fontweight="bold"
    )

    plt.figtext(
        0.63, 0.56,
        "Manchester City",
        ha="center",
        va="center",
        fontsize=6.2,
        color=city_color,
        fontweight="bold"
    )

    # ----------------------------
    # Percentages
    # ----------------------------
    plt.figtext(
        0.37, 0.30,
        f"{arsenal_val}%",
        ha="center",
        va="center",
        fontsize=12,
        fontweight="bold",
        color=arsenal_color
    )

    plt.figtext(
        0.63, 0.30,
        f"{city_val}%",
        ha="center",
        va="center",
        fontsize=12,
        fontweight="bold",
        color=city_color
    )

    # ----------------------------
    # Save
    # ----------------------------
    file_path = os.path.join(output_dir, filename)

    plt.savefig(
        file_path,
        dpi=300,
        bbox_inches="tight",
        pad_inches=0,
        facecolor="white"
    )

    plt.show()
    plt.close()


# ----------------------------
# Create cards
# ----------------------------
create_big_card("If Arsenal Win", 99, 1, "arsenal_win_card.png")
create_big_card("If Draw", 90, 10, "draw_card.png")
create_big_card("If Manchester City Win", 65, 35, "city_win_card.png")
```

```python
import matplotlib.pyplot as plt
import os
from matplotlib.patches import Patch

# ----------------------------
# DATA
# ----------------------------
matchweeks = ["Matchday\n31", "Matchday\n34", "Matchday\n35",
              "Matchday\n36", "Matchday\n37", "Matchday\n38"]

arsenal = [
    None,
    "Newcastle\n(14)",
    "Fulham\n(12)",
    "West Ham\n(17)",
    "Burnley\n(19)",
    "Crystal Palace\n(13)"
]

city = [
    "Crystal Palace\n(13)",
    "Burnley\n(19)",
    "Everton\n(10)",
    "Brentford\n(7)",
    "Bournemouth\n(8)",
    "Aston Villa\n(4)"
]

arsenal_home = [None, True, True, False, True, False]
city_home    = [True, False, False, True, False, True]

data = [arsenal, city]
home_flags = [arsenal_home, city_home]

# ----------------------------
# LAYOUT
# ----------------------------
col_spacing = 0.9
left_offset = 0.6

row_height = 0.85
row_gap = 0.18

n_rows = 2
n_cols = len(matchweeks)

total_height = n_rows * (row_height + row_gap)

fig, ax = plt.subplots(figsize=(8.2, 3.0), dpi=200)

fig.patch.set_facecolor("white")
ax.set_facecolor("white")

arsenal_color = "#EF0107"
city_color = "#6CABDD"

# ----------------------------
# AXIS
# ----------------------------
ax.set_xlim(0, left_offset + col_spacing * n_cols + 0.2)
ax.set_ylim(0, total_height + 0.25)
ax.axis("off")

# ----------------------------
# TITLE
# ----------------------------
fig.text(
    0.5, 0.965,
    "Premier League Title Race Run-In",
    ha="center",
    fontsize=13,
    fontweight="bold"
)

fig.lines.append(plt.Line2D(
    [0.18, 0.82], [0.935, 0.935],
    transform=fig.transFigure,
    color="#dddddd",
    linewidth=1
))

fig.text(
    0.5, 0.89,
    "Arsenal and Manchester City remaining fixtures, including opposition strength (league position) and venue advantage",
    ha="center",
    fontsize=9,
    color="#666666"
)

# ----------------------------
# GRID
# ----------------------------
ax.plot(
    [left_offset - 0.2, left_offset - 0.2],
    [0, total_height],
    color="#dddddd",
    lw=0.8
)

for row in range(n_rows):
    for col in range(n_cols):

        x = left_offset + col * col_spacing
        y = total_height - (row + 1) * (row_height + row_gap) + row_gap

        value = data[row][col]
        is_home = home_flags[row][col]

        # ----------------------------
        # SKIP EMPTY FIXTURES COMPLETELY
        # ----------------------------
        if value is None:
            continue

        bg = "#eef6ff" if is_home else "#fff2f2"

        rect = plt.Rectangle(
            (x, y),
            col_spacing * 0.95,
            row_height,
            facecolor=bg,
            edgecolor="#eeeeee",
            linewidth=0.6
        )
        ax.add_patch(rect)

        ax.text(
            x + col_spacing * 0.475,
            y + row_height / 2,
            value,
            ha="center",
            va="center",
            fontsize=8.5,
            color="#222222"
        )

# ----------------------------
# TEAM LABELS
# ----------------------------
ax.text(
    left_offset - 0.35,
    total_height - (row_height + row_gap) / 2,
    "Arsenal",
    ha="right",
    va="center",
    fontsize=10,
    fontweight="bold",
    color=arsenal_color
)

ax.text(
    left_offset - 0.35,
    total_height - 1.5 * (row_height + row_gap),
    "Manchester\nCity",
    ha="right",
    va="center",
    fontsize=10,
    fontweight="bold",
    color=city_color
)

# ----------------------------
# HEADERS
# ----------------------------
header_y = total_height + 0.08

for i, mw in enumerate(matchweeks):
    x = left_offset + i * col_spacing + col_spacing * 0.475

    ax.text(
        x,
        header_y,
        mw,
        ha="center",
        va="bottom",
        fontsize=7.5,
        color="#666666"
    )

# ----------------------------
# FOOTER
# ----------------------------
ax.text(
    0.96, 0.00,
    "Victoria Friss de Kereki",
    ha="right",
    va="bottom",
    fontsize=7.5,
    color="#888888",
    transform=ax.transAxes
)

ax.text(
    0.96, -0.045,
    "Sports Analytics & Data Science",
    ha="right",
    va="bottom",
    fontsize=6.8,
    color="#aaaaaa",
    transform=ax.transAxes
)

# ----------------------------
# LEGEND
# ----------------------------
legend_elements = [
    Patch(facecolor="#eef6ff", label="Home fixture"),
    Patch(facecolor="#fff2f2", label="Away fixture")
]

fig.legend(
    handles=legend_elements,
    loc="lower center",
    bbox_to_anchor=(0.5, 0.035),
    ncol=2,
    frameon=False,
    fontsize=9
)

# ----------------------------
# SPACING
# ----------------------------
plt.subplots_adjust(left=0.05, right=0.98, top=0.81, bottom=0.12)

# ----------------------------
# SAVE
# ----------------------------
output_dir = "Images and others"
os.makedirs(output_dir, exist_ok=True)

plt.savefig(
    os.path.join(output_dir, "fixtures_arsenal_mancity.png"),
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# ----------------------------
# DATA
# ----------------------------
teams = [
    "Arsenal",
    "Manchester City",
    "Manchester United",
    "Aston Villa",
    "Liverpool",
    "Chelsea"
]

top5 = [100, 100, 99, 98, 93, 5]

# ----------------------------
# STYLE
# ----------------------------
plt.style.use("seaborn-v0_8-white")

fig, ax = plt.subplots(figsize=(6.2, 3.2), dpi=200)

fig.patch.set_facecolor("#ffffff")
ax.set_facecolor("#ffffff")

arsenal_color = "#EF0107"
city_color = "#6CABDD"

# ----------------------------
# LAYOUT
# ----------------------------
y = np.arange(len(teams))[::-1]
row_h = 0.68

ax.set_xlim(0, 61)
ax.set_ylim(-0.5, len(teams) - 0.5)

# ----------------------------
# ROWS
# ----------------------------
for i, (team, prob) in enumerate(zip(teams, top5)):
    yy = y[i]

    ax.add_patch(plt.Rectangle(
        (0, yy - row_h/2),
        61,
        row_h,
        facecolor="#f8fafc" if i % 2 == 0 else "#ffffff",
        edgecolor="#e5e7eb",
        linewidth=0.6
    ))

    if team == "Arsenal":
        color = arsenal_color
        weight = "bold"
    elif team == "Manchester City":
        color = city_color
        weight = "bold"
    else:
        color = "#374151"
        weight = "normal"

    ax.text(2, yy, team,
            va="center", ha="left",
            fontsize=9.2,
            color=color,
            fontweight=weight)

    ax.text(60, yy, f"{int(round(prob))}%",
            va="center", ha="right",
            fontsize=9.2,
            color="#111827",
            fontweight="bold")

# ----------------------------
# TITLE
# ----------------------------
fig.text(
    0.5, 0.965,
    "Premier League Top 5 Probability",
    ha="center",
    fontsize=12.5,
    fontweight="bold"
)

fig.lines.append(plt.Line2D(
    [0.18, 0.82], [0.94, 0.94],
    transform=fig.transFigure,
    color="#dddddd",
    linewidth=1
))

fig.text(
    0.5, 0.895,
    "Sum of finishing probabilities from 10,000 Monte Carlo simulations",
    ha="center",
    fontsize=9,
    color="#666666"
)

# ----------------------------
# CLEAN AXES
# ----------------------------
ax.axis("off")

# ----------------------------
# 🔥 FOOTER (NOW PROPERLY BELOW EVERYTHING)
# ----------------------------
fig.text(
    0.9, 0.095,
    "Victoria Friss de Kereki",
    ha="right",
    fontsize=8.2,
    color="#888888"
)

fig.text(
    0.9, 0.065,
    "Sports Analytics & Data Science",
    ha="right",
    fontsize=7.5,
    color="#aaaaaa"
)

# ----------------------------
# SAVE (reserve bottom space so footer doesn't overlap)
# ----------------------------
output_dir = "Images and others"
os.makedirs(output_dir, exist_ok=True)

plt.savefig(
    os.path.join(output_dir, "top5_probability.png"),
    dpi=300,
    facecolor="white",
    bbox_inches="tight"
)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# ----------------------------
# RAW RELEGATION DATA
# ----------------------------
data = {
    "Wolverhampton Wanderers": 100,
    "Burnley": 100,
    "Tottenham Hotspur": 58,
    "West Ham United": 39,
    "Nottingham Forest": 3,
    "Leeds United": 0.3  # <1%
}

# ----------------------------
# SORT
# ----------------------------
sorted_data = sorted(data.items(), key=lambda x: x[1], reverse=True)

teams = [x[0] for x in sorted_data][::-1]
relegation = [x[1] for x in sorted_data][::-1]

# ----------------------------
# STYLE
# ----------------------------
plt.style.use("seaborn-v0_8-white")

fig, ax = plt.subplots(figsize=(6.2, 3.2), dpi=200)

fig.patch.set_facecolor("#ffffff")
ax.set_facecolor("#ffffff")

# ----------------------------
# RED THEME (NO CONDITIONAL FORMATTING)
# ----------------------------
light_red = "#fef2f2"

# ----------------------------
# LAYOUT
# ----------------------------
row_h = 0.68

y = np.arange(len(teams))[::-1]

ax.set_xlim(0, 61)
ax.set_ylim(-0.5, len(teams) - 0.5)

# ----------------------------
# ROWS
# ----------------------------
for i, (team, prob) in enumerate(zip(teams, relegation)):
    yy = y[i]

    ax.add_patch(plt.Rectangle(
        (0, yy - row_h / 2),
        61,
        row_h,
        facecolor=light_red if i % 2 == 0 else "#ffffff",
        edgecolor="#e5e7eb",
        linewidth=0.6
    ))

    # TEAM NAME (UNIFORM)
    ax.text(
        2,
        yy,
        team,
        va="center",
        ha="left",
        fontsize=9.2,
        color="#374151",
        fontweight="normal"
    )

    # VALUE (UNIFORM - NO CONDITIONAL FORMATTING)
    ax.text(
        60,
        yy,
        f"{int(round(prob))}%" if team != "Leeds United" else "<1%",
        va="center",
        ha="right",
        fontsize=9.2,
        color="#111827",
        fontweight="bold"
    )

# ----------------------------
# TITLE
# ----------------------------
fig.text(
    0.5, 0.965,
    "Premier League Relegation Probability",
    ha="center",
    fontsize=12.5,
    fontweight="bold",
    color="#111111"
)

fig.lines.append(plt.Line2D(
    [0.18, 0.82], [0.94, 0.94],
    transform=fig.transFigure,
    color="#e5e7eb",
    linewidth=1
))

fig.text(
    0.5, 0.895,
    "Top teams most likely to finish in relegation zone (positions 18-20)",
    ha="center",
    fontsize=9,
    color="#666666"
)

# ----------------------------
# CLEAN AXES
# ----------------------------
ax.axis("off")

# ----------------------------
# FOOTER (BOTTOM ALIGNED)
# ----------------------------
fig.text(
    0.9, 0.095,
    "Victoria Friss de Kereki",
    ha="right",
    fontsize=8.2,
    color="#888888"
)

fig.text(
    0.9, 0.065,
    "Sports Analytics & Data Science",
    ha="right",
    fontsize=7.5,
    color="#aaaaaa"
)

# ----------------------------
# SAVE
# ----------------------------
output_dir = "Images and others"
os.makedirs(output_dir, exist_ok=True)

plt.savefig(
    os.path.join(output_dir, "relegation_probability.png"),
    dpi=300,
    facecolor="white",
    bbox_inches="tight"
)

plt.show()
```
